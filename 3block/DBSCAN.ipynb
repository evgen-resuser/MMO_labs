**DBSCAN** — это алгоритм кластеризации, основанный на плотности, который не требует заранее задавать количество кластеров и может находить кластеры произвольной формы, а также выделять шум

**Основные параметры**
- eps (ε) — максимальное расстояние между двумя точками, чтобы считать их соседями
- min_samples — минимальное количество точек в окрестности для формирования кластера

**Типы точек**:

- Ядровая (core) — если в радиусе eps есть >= min_samples точек
- Граничная (border) — сама не ядро, но находится рядом с ядром
- Шум (noise) — не принадлежит ни одному кластеру
  
**Алгоритм**:
1) Берем произвольную точку
2) Если она ядровая, то создаём новый кластер
3) Добавляем все точки в её эпсилон-окрестности
4) Рекурсивно расширяем кластер через соседние ядровые точки
5) Если точка не ядровая и не принадлежит кластеру, то она помечается как шум
6) Повторяем, пока не обработаем все точки

**Плюсы**:
- Не нужно заранее задавать число кластеров
- Находит кластеры произвольной формы
- Хорошо работает с шумом (явно его выделяет)

**Минусы**:
- Чувствителен к выбору параметров ε и MinPts
- Плохо работает при разной плотности кластеров
- В высоких размерностях теряет эффективность

```python
# Блок с используемыми библиотеками
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# Блок с описанием и демонстрацией
df = pd.read_csv("../datasets/spotify_tracks.csv", sep=",")
df.head()
```

```python
# Блок с предварительной обработкой датасета

df.info()

# заменим жанры на категории

df_genre = df.copy()
df_genre = pd.get_dummies(df, columns=['genre'], drop_first=True)

df_tr = df_genre
df_tr.info()
```

```python
X = df_tr.drop(['track_id', 'track_name', 'artist_name', 'album_name', 'release_year', 'time_signature', 'duration_ms', 'popularity'], 
               axis=1) 

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=2)
pca_result = pca.fit_transform(X_scaled)

# Визуализация
plt.figure(figsize=(10, 8))
plt.scatter(pca_result[:, 0], pca_result[:, 1], alpha=0.6)
plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.2%})')
plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.2%})')
plt.title('PCA Visualization')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
dbscan = DBSCAN(eps=0.16, min_samples=60)
clusters = dbscan.fit_predict(pca_result)

plt.figure(figsize=(10, 8))
scatter = plt.scatter(pca_result[:, 0], pca_result[:, 1], 
                      c=clusters, cmap='viridis', alpha=0.6)
plt.colorbar(scatter, label='Cluster ID')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('DBSCAN Clustering')
```

```python
score = silhouette_score(X_scaled, clusters)
print("Silhouette Score:", score)
```
